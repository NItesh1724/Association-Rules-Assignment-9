Creating frequent datasets

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemset=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemset
```

Extrating frequent itemset from the association rule

```python
from mlxtend.frequent_patterns import association_rules
```

```python
assoc_rule=association_rules(frequent_itemset,min_threshold=1.0)
assoc_rule
```

```python
assoc_rule[assoc_rule['lift']>1]
```

```python
plt.scatter(assoc_rule['support'],assoc_rule['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
assoc_rule[assoc_rule['lift']>2]
```

```python
assoc_rule1=association_rules(frequent_itemset,metric='lift',min_threshold=0.9)
```

```python
assoc_rule1[assoc_rule1['lift']>1]
```

```python
plt.scatter(assoc_rule1['support'],assoc_rule1['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
frequent_itemset.sort_values('support',ascending=False)
```
